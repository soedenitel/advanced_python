```python
import pandas as pd


firefighters_df = pd.read_csv('https://video.ittensive.com/python-advanced/data-5283-2019-10-04.utf.csv', sep=';')
unemployed_df = pd.read_csv('https://video.ittensive.com/python-advanced/data-9753-2019-07-25.utf.csv', sep=';')
```

```python
firefighters_df = firefighters_df.set_index(['Month', 'Year'])
firefighters_df.head()
```

```python
unemployed_df.rename(columns={'Period': 'Month'}, inplace=True)
unemployed_df = unemployed_df.set_index(['Month', 'Year'])
unemployed_df.head()
```

Объедините эти данные индексами (Месяц/Год) с данными из предыдущего задания (вызовы пожарных) для Центральный административный округ:

```python
united_df = pd.merge(firefighters_df, unemployed_df, left_index=True, right_index=True)
```

Найдите значение поля UnemployedMen в том месяце, когда было меньше всего вызовов в Центральном административном округе.

```python
min_value = united_df[united_df['AdmArea'] == 'Центральный административный округ']['Calls'].min()
filter_series = united_df[united_df['AdmArea'] == 'Центральный административный округ']['Calls'] == min_value
filter_series.name='filter_field'
tmp_df = pd.merge(filter_series, united_df, left_index=True, right_index=True)
tmp_df[tmp_df['filter_field']]['UnemployedMen']
```
