# Задание: многостраничный отчет

Используя данные по активностям в парках Москвы

https://video.ittensive.com/python-advanced/data-107235-2019-12-02.utf.json

Создайте PDF отчет, в котором выведите:
1. Диаграмму распределения числа активностей по паркам, топ10 самых активных
2. Таблицу активностей по всем паркам в виде Активность-Расписание-Парк

```python
# Отключим предупреждения
import warnings
warnings.simplefilter('ignore')

import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib import rcParams
rcParams['figure.figsize'] = (10, 7)

df = pd.DataFrame(
    json.loads(requests.get('https://video.ittensive.com/python-advanced/data-107235-2019-12-02.utf.json').content),
)
df = df.set_index('ID')
df.head()
```

## Диаграмма распределения числа активностей по паркам, топ10 самых активных

```python
park_dict = dict()
for park in df['NameOfPark']:
    park_dict[park['global_id']] = park['value']

df['park_id'] = df['NameOfPark'].map(lambda park: park['global_id'])
top_10 = df['park_id'].value_counts()[:10]
labels = [park_dict[park_id] + ': ' + str(top_10[park_id]) for park_id in top_10.index]
top_10.plot(kind='pie', labels=labels, title='Число активностей по топ-10 паркам', label='', fontsize='7')
plt.savefig('homework_3.png')
```

## Таблица активностей по всем паркам в виде Активность-Расписание-Парк

```python
activity_df = df[['CourseName', 'CoursesTimetable']]
activity_df['Park'] = df['park_id'].map(lambda park_id: park_dict.get(park_id))
activity_df.rename(columns={'CourseName': 'Активность', 'CoursesTimetable': 'Время проведения', 'Park': 'Место проведения'}, inplace=True)
activity_df.set_index('Активность', inplace=True)
pd.set_option('display.max_colwidth', 1000)
activity_df
```

## Создайте PDF отчет

```python
import pdfkit
from jinja2 import Template
import binascii

options = {
    'page-size': 'A4',
    'header-right': '[page]'
}
config = pdfkit.configuration(wkhtmltopdf='C:/Program Files/wkhtmltopdf/bin/wkhtmltopdf.exe')

with open('homework_3.png', 'rb') as img_file:
    img = 'data:image/png;base64,' + binascii.b2a_base64(img_file.read(), newline=False).decode('UTF-8')

html_template = '''
<html>
    <head>
        <title>Таблица активностей по паркам</title>
        <meta charset="utf-8"/>
        <style>
            th {
                text-align: center;
            }
            td {
                text-align: left;
            }
        </style>
    </head>
    <body>
        <h1>ТОП-10 парков</h1>
        <img src="{{data.image}}" alt="ТОП-10 парков">
        <h2>Данные по всем паркам</h2>
        {{data.table}}
    </body>
</html>
'''

html = Template(html_template).render(
    data={
        'image': img,
        'table': activity_df.to_html()
    }
)

pdfkit.from_string(html, 'homework_3.pdf', options=options, configuration=config)
```

## Сколько активностей Тайцзицюань есть в парках Москвы?

```python
print('Всего {0:d} активность(ей) Тайцзицюань есть в парках Москвы'.format(
    activity_df.index.str.contains('Тайцзицюань').sum()
))
```
